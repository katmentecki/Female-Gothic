```python
df = pd.read_csv("female_gothic_complete.csv", encoding="latin-1")
ids = list(df.id)
titles = list(df.title)
ids_dict = {}
for i in ids:
    ids_dict[i] = True
```

```python
df2 = pd.read_csv("male_gothic_complete.csv", encoding="latin-1")
mids = list(df2.id)
male_titles = list(df2.title)
male_ids_dict = {}
for i in mids:
    male_ids_dict[i] = True
```

```python
import os
import json
import glob
    
def look_for_ids(folder, ids_dict, destination):
    #look in volumes
    for root, dirs, files in os.walk(folder, topdown=False):
        for name in files:
            if "json" in name:
                with open(os.path.join(root, name))as f:
                    myjson = json.loads(f.read())
                    #go through each json
                    work_id = myjson['work_id']
                    #see if it matches an id in ids
                    try:
                        ids_dict[work_id]
                        #if it does, we get txt, add to a folder of txts, title will equal id
                        #look in root for a txt file
                        txt_file = glob.glob(root+"/*.txt")[0]
                        with open (txt_file) as t:
                            full = t.read()
                        newfilename = destination+"/"+str(work_id)+".txt"
                        with open(newfilename, "a") as new:
                            #will join to multivolume if match previous (append)
                            new.write(full)
                        new.close()
                    except:
                        pass
    
    
```
